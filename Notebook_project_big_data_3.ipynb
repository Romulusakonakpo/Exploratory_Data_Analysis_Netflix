## Introduction

Since 1997, __Netflix__, the video streaming giant, has revolutionized the way we consume audiovisual content. With its vast catalog of movies and TV shows, Netflix has captivated millions of subscribers around the world. But behind this façade of entertainment lies a wealth of valuable information about movie and TV genres, the most popular actors, and user preferences.
It is in this context that we invite you to dive into the world of Netflix records through an __EXPLORATION, CLEANSING AND VISUALISATION__ project.

### Project's subject

Netflix is one the world's leading entertainment services with over 260 millions paid memberships in over 190 countries enjoying TV series, films and games across a wide variety of genres and languages. Members can play, pause and resume watching as much as they want, anytime, anywhere, and can change their plans at any time.

### Features

We'll leverage public datasets from Kaggle and UNESCO's Institute of Statistics. The [Kaggle dataset](https://www.kaggle.com/datasets/nayanack/netflix/data), containing information on movies and TV shows on Netflix (titles, genres, directors, etc.), will be our primary source for exploration.
For the first dataset from __Kaggle__, it contains information about the movies and TV shows available on Netflix, including titles, genres, directors, actors, and release dates.
Number of Entries: The dataset contains 8807 entries.
Columns: There are 12 columns in total.
Here is the details of the name of each column.
- show_id : Unique identifier for each show.
- type : Indicates whether the entry is a movie or a TV show.
- title : The title of the movie or TV show.
- director : The director(s) of the movie or TV show.
- cast : The cast members of the movie or TV show.
- country : The country or countries where the movie or TV show was produced.
- date_added : The date when the movie or TV show was added to Netflix.
- release_year : The year when the movie or TV show was released.
- rating : The rating assigned to the movie or TV show.
- duration : The duration of the movie or TV show.
- listed_in : The categories or genres that the movie or TV show belongs to.
This dataset can be used for various analytical purposes such as exploring trends in Netflix content, analyze the distribution of content origin (countries) for shows and movies, identify the types of shows (TV shows vs movies) and explore genre distribution, examine director participation and identify top contributors, and more.

For the [second dataset](http://data.uis.unesco.org/) from __UNESCO__, the Institute of Statistics conducts a biennial survey on feature film and cinema statistics to monitor global trends in selected areas of this industry. This is the only international survey to include countries in all regions and at all stages of development from Hollywood to Bollywood and Nollywood. It uses comparable definitions to collect data on a range of issues, including the Top 10 films viewed per country; linguistic diversity in film production; admissions; national versus international co-productions per country; and distribution mechanisms.

The data can be used to track trends in consumption, for example, by looking at the origins of films viewed versus the most popular films, based on attendance. The data also reflect issues related to cinema infrastructure and access, as reflected by the numbers of indoor cinemas and average ticket prices per capita.

### Hypotheses of the project

Our goal is to utilize data visualization techniques to test several hypotheses, such as :

*H1 : The first film released on Netflix dates from 1997 when the platform was created.*

*H2 : The US and India are in the top 5 of countries which release the most shows and movies on Netflix.*

*H3 : The US and India release are in the top 5 of countries which release the most shows and movies in the world.*

*H4 : The countries that spend the most in the cinema industry are those that produce the most films.*

*H5 : There are more TV shows than movies on Netflix.*

*H6 : Some genres dominate the Netflix library, with action, comedy, and drama being the most popular.*

*H7 : The films on Netflix are more aimed at adults than other categories of people.*

*H8 : A small number of directors contribute significantly to the content on Netflix.*

*H9 : The director who produces the most films on Netflix comes from the country which releases the most films.*

*H10 : Netflix can be used for educational purposes by offering documentaries on various subjects.*

### Data exploration

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df['rating'].unique()
```

```python
# Recoding of the modalities of 'rating' 
replacement_dict = {
    'PG-13': 'Teens(13+)',
    'PG': 'Parental Guidance Suggested',
    'TV-MA': 'Adults(18+)',
    'TV-14': 'Young Adults(14+)',
    'TV-PG': 'Older Kids(8+) Parental Guidance Suggested',
    'TV-Y': 'Kids(All)',
    'TV-Y7': 'Older Kids(7+)',
    'R': 'Adults only',
    'NC-17': 'Adults(17+) and Warning Fantasy Violence',
    'TV-G': 'Family(All) Suitable for all ages',
    'TV-Y7-FV': 'Older Kids(7+) and Warning Fantasy Violence',
    'NR': 'NotRated',
    'G': 'Kids(All ages)',
    'UR': 'Unrated'
}
df['rating'] = df['rating'].replace(replacement_dict)
```

```python
# Removing unwanted values
remove_dict= {'74 min': np.nan, 
              '84 min': np.nan, 
              '66 min': np.nan}
df['rating']=df['rating'].replace(remove_dict) 
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors = 'coerce')
```

```python
a={'show_id': 'string',
 'type': 'category',
 'title': 'string',
 'director': 'string',
 'cast': 'string',
 'country': 'category',
  'rating': 'category',
 'duration': 'category',
 'listed_in': 'string',
 'description': 'string'}
df=df.astype(a)
```

### Data visualisation

```python
plt.pie(df['type'].value_counts(),labels=['Movie', 'TV Show'], explode=(0,0.1), startangle=90,shadow=True,autopct='%.0f%%')
plt.show()
```

From this graphic, we see that there are more Movies (70%) than TV shows (30%) on Netflix.

```python
df.groupby('rating', observed=True)['title'].count().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='rating',data=df,order=df['rating'].value_counts().index[:])
plt.xticks(rotation=90)
plt.show();
```

We can deduce from this graph that most of the films on Netflix are for the 'Adults (18+)' category.

```python
df=df.assign(country=df['country'].str.split(',')).explode('country')
df['country']=df['country'].str.lstrip()
df.replace({'': np.nan}, inplace=True)
df['country'].dropna(inplace=True)
df = df.sort_values('country', ascending=True)
df.reset_index(drop=True, inplace=True)
```

```python
gb = df.groupby('country',observed=True)['type'].agg(['count'])
gb_sort = gb.sort_values(by='count', ascending=False)
gb_sort.head(15)
```

```python
plt.figure(figsize=(25,25))
sns.countplot(x='country',data=df,order=df['country'].value_counts().index[:15])
plt.xticks(rotation=90)
plt.show()
```

From this graph, we can see that the country that produces the most visual content on Netflix is United States. India comes second.

```python
plt.figure(figsize=(15,5))
set=sns.countplot(x=df['listed_in'],order=df['listed_in'].value_counts(ascending=False).index[:15])
set.set_xticklabels(set.get_xticklabels(),rotation=90)
plt.show();
```

From this graph, it appears that Netflix broadcasts as many documentaries as drama genre content. Action films, adventure films or even romantic films do not take precedence on the platform. It can be used for educational purposes.

```python
df=df.assign(director=df['director'].str.split(',')).explode('director')
df['director']=df['director'].str.lstrip()
df.replace({'': np.nan}, inplace=True)
df['director'].dropna(inplace=True)
```

```python
df.groupby('director')['title'].count().sort_values(ascending=False).head(14)
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='director',data=df,order=df['director'].value_counts().index[:15])
plt.xticks(rotation=90)
plt.show()
```

From this graph, it is easy to see that the director who contributes the most to content on Netflix is Rajiv Chilaka. According to our research, the latter is an Indian Creator, Director. He is the founder and CEO of Green Gold Animations based in Hyderabad, India.
